## Posiciones en el ranking para algunos modelos

```python
res_order = results.sort_values(by=['element_key','win_mean']).reset_index(drop=True)
```

```python
models = sorted(res_order.model.unique())
```

```python
col_names =  ['modelo', 'ranking_medio']
ranking  = pd.DataFrame(columns = col_names)

for mod in models:
    rank = np.average(np.array(res_order.loc[res_order.model == mod].index)-np.array(range(0,23*30,23)))
    ranking.loc[len(ranking)] = [mod, rank]

ranking
```
